```python
import pandas as pd
import matplotlib.pyplot as plt #Libraries for visualization
import seaborn as sns
import numpy as np #Library for numerical calculations
from scipy import stats #Library for statistical analysis
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler #Min max Scaler for normalization of data to range [0,1]
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet, BayesianRidge #Importing the linear regression models
from sklearn.metrics import mean_squared_error #For evaluation of the models
```

```python
df_housing = pd.read_csv("../data/california-housing/california-housing.csv")
print(df_housing)
```

```python
median_h_val = df_housing["median_house_value"]
features = df_housing.drop(["median_house_value"],axis=1)
number_of_features = df_housing.shape[1]
feature_names = tuple(df_housing.columns[:-1])
print("Number of features:",number_of_features)
print("Feature names:", feature_names)
N = df_housing.shape[0]
print("Number of examples", N)
```

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
fig = plt.figure(figsize=(10, 10))
fig.subplots(number_of_features//2+1, ncols=2)
fig.tight_layout(pad=6)
for feat_i in range(number_of_features): #The histograms of the features
    ax = plt.subplot(number_of_features//2+1,2, feat_i+1)
    plt.title(feature_names[feat_i])
    sns.distplot(features[feature_names[feat_i]].values, color = "navy") 
ax = plt.subplot(number_of_features//2+1,2, feat_i+2)
plt.title("Target: " + feature_names[0]) 
sns.distplot(median_h_val.values, color = "darkorange") #The histograms of the target variable
plt.show()
```

```python
for feat_i in range(number_of_features): 
    plt.figure()
    plt.scatter( features[feature_names[feat_i]].values, median_h_val.values,  alpha=.8, color="darkorange") 
    plt.xticks(())
    plt.yticks(())
    ax.axis('equal')
    plt.title(feature_names[feat_i] + " x " + df_housing.columns[-1])
    plt.show()
```

```python
for feat_i in range(number_of_features):
    sns.jointplot(features[feature_names[feat_i]].values, median_h_val.values, kind="kde", space=0, color="g")
plt.show()
```

```python
correlation_matrix = np.ones((number_of_features +1 ,number_of_features +1))
for i in range(number_of_features): #The first 3x3 part of the matrix represents the feature correlations
    measure = features[feature_names[i]].values
    for j in range(number_of_features):
        measure2 = features[feature_names[j]].values
        corr, _ = stats.pearsonr(measure, measure2)
        correlation_matrix[i][j] = corr
for i in range(number_of_features): #The last column and row represent the feature to target variable correlations
    j = number_of_features
    measure = features[feature_names[i]].values
    measure2 = median_h_val.values
    corr, _ = stats.pearsonr(measure, measure2)
    correlation_matrix[i][j] = corr
    correlation_matrix[j][i] = corr
plt.figure()
plt.imshow(correlation_matrix, cmap = "inferno", vmin=-1, vmax=1) #We can draw the matrix using imshow
plt.colorbar()
plt.show()
```

**MIN MAX NORMALIZATION**

```python
scaler = MinMaxScaler()
X = scaler.fit_transform(features.values)
X_train, X_test, y_train, y_test = train_test_split(X, median_h_val.values, test_size=0.30, random_state=42) 
```

```python
reg = LinearRegression().fit(X_train, y_train)
print("Coefficients W")
print(reg.coef_)
print("Intercept W0")
print(reg.intercept_ )
y_pred = reg.predict(X_test)
print("Mean Squared Error")
print(mean_squared_error(y_test, y_pred))

plt.figure()
plt.plot(np.arange(len(reg.coef_)), reg.coef_, ".",  markersize=15, color = "darkgreen")
plt.xlabel("Feature number")
plt.ylabel("Coefficients w")
plt.title("Coefficient Weights")
plt.show()
```
